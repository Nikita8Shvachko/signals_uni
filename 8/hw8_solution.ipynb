Задача 1
--------
Для мультиплитивной группы поля $GF(5^2)$, считая $x$ - примитивным элементом, определить порядки всех элементов поля.

Задача 2
--------
Для кодера $(15, 4)$-кода с проверочным полиномом $h(x) = 1 + x + x^4$, описать смену состояний регистра при начальном заполнении $0, 0, 1, 1$.
- убедиться, что это код максимальной длины
- определить минимальное расстояние этого кода

```python
from math import gcd # библиотека для вычисления НОД


def task1_orders_gf_5_2():  # Порядки элементов GF(5^2)^* при условии, что x — примитивный.
    n = 5**2 - 1  # 24

    print("Задача 1")
    print("Мультипликативная группа GF(5^2)^* циклическая, имеет порядок 24.")
    print("Пусть x - примитивный элемент, тогда ord(x)=24.")
    print("Элементы поля: 0 и x^0, x^1, ..., x^23 (причем x^0=1).")
    print()

    print("Порядки элементов поля:")
    print("  0: порядок не определен (нет условия a^s=1 для мультипликативной группы).")
    for k in range(24):  # 24 - порядок мультипликативной группы GF(5^2)^*
        order = n // gcd(n, k)  # порядок элемента x^k
        elem = "1" if k == 0 else f"x^{k}"
        print(f"  {elem:>4}: {order}")  # выводит порядок элемента x^k


def lfsr_states_poly_h(start_state, steps=15):
    # LFSR для h(x)=1+x+x^4.

    # Считаем состояние как (m3, m2, m1, m0).
    # По конспекту: выход = m3, сдвиг (m3<-m2, m2<-m1, m1<-m0).
    # Обратная связь: m0_new = m3 XOR m2.

    states = [tuple(start_state)]  # состояния регистра
    outputs = []  # выходные биты
    s = tuple(start_state)  # текущее состояние регистра

    for _ in range(steps):  # цикл по количеству шагов
        m3, m2, m1, m0 = s  # извлекаем биты состояния
        outputs.append(m3)  # добавляем выходной бит в список
        new_m0 = m3 ^ m2  # вычисляем новый бит m0
        s = (m2, m1, m0, new_m0)  # обновляем состояние регистра
        states.append(s)  # добавляем новое состояние в список

    return states, outputs  # возвращаем списки состояний и выходных бит


def task2_lfsr_and_distance():
    print("Задача 2")
    h_desc = "h(x)=1 + x + x^4"
    print(f"Проверочный полином: {h_desc}")

    start_state = (0, 0, 1, 1)  # (m3, m2, m1, m0)
    states, outputs = lfsr_states_poly_h(
        start_state=start_state, steps=15
    )  # вычисляем состояния и выходные биты

    print()
    print("Смена состояний регистра (m3,m2,m1,m0) и выходных битов:")
    for t in range(15):  # цикл по количеству шагов
        m3, m2, m1, m0 = states[t]  # извлекаем биты состояния
        nxt = states[t + 1]  # следующее состояние
        print(
            f"  t={t:2d}: ({m3},{m2},{m1},{m0}) -> выход {m3} -> ({nxt[0]},{nxt[1]},{nxt[2]},{nxt[3]})"
        )

    output_seq = "".join(map(str, outputs))  # преобразуем список выходных бит в строку
    print()
    print(f"Выходная последовательность (длина 15): {output_seq}")

    # Проверка максимальной длины: цикл проходит через все 2^4-1 ненулевых состояния.
    visited = set(states[:-1])  # множество посещенных состояний (кроме последнего)
    max_len = 2**4 - 1  # максимальная длина
    is_cycle = states[-1] == start_state  # проверка, вернулся ли в начальное состояние

    print()
    print("Проверка максимальной длины:")
    print(
        f"  Прошло различных состояний (кроме повторения): {len(visited)} (должно быть {max_len})"
    )
    print(
        f"  После 15 шагов вернулись в начальное состояние: {'да' if is_cycle else 'нет'}"
    )
    print(
        "  Значит, период равен 15 и регистр проходит через все 2^4-1 ненулевых состояний."
    )

    # По конспекту для кода максимальной длины (двойственный к коду Хэмминга):
    # при m=4: n=2^m-1=15, d_min = 2^(m-1) = 8.
    d_min = 2 ** (4 - 1)  # минимальное расстояние
    k = 4  # размерность кода
    n = 15  # длина кода

    print("Минимальное расстояние кода:")
    print(f"  Параметры кода: (n,k)=({n},{k})")
    print(f"  d_min = 2^(m-1) = 2^3 = {d_min}")


task1_orders_gf_5_2()
print("\n" + "=" * 60 + "\n")
task2_lfsr_and_distance()
```
